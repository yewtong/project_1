````python
# to check sat_2018 for errors
# def a function that can check all subsq SAT datasets

def check_sat(dataframe):
    
    # check for duplicate/missing States
    print("There are {} States in the dataset.".format(dataframe["State"].nunique()))
    
    # check that participation rates are between 0-100%
    for rate in dataframe["Participation"]:
        if rate<0 or rate>100:
            print(f"{rate}% participation rate is out of range.")
    print("Check is completed. Unlisted participation rates are in range between 0-100%.")
    
    # check test scores are between 200-800
    cols = ["Evidence-Based Reading and Writing", "Math"]
    for col in cols:
        for score in dataframe[col]:
            if score<200 or score>800:
                print(f"{score} score for {col} is out of range.")
        print(f"Check is completed. Unlisted {col} test scores are in range between 200-800.")
    
    # check total test scores are between 400-1600
    for score in dataframe["Total"]:
        if score<400 or score>1600:
            print(f"{score} total score is out of range.")
    print("Check is completed. Unlisted total test scores are in range between 400-1600.")   
````

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
# popn std = sqr(sum[(popn[i] - popn.mean()).squared + (popn[i+1] - popn.mean()).squared + ...] / popn size)
````

````python
def popn_std(series):
    dist_sq = [] # create empty list for square of dist from mean
    mean = series.mean() # eval mean of series
    for s in series:
        dist_sq.append((s-mean)**2) # append to list square of dist from mean
    var = sum(dist_sq) / len(series) # calc variance by dividing sum of list of squared dist from mean by total popn size
    std = np.sqrt(var) # eval std with the square-root of variance
    return std
````

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels):
    nrows = int(np.ceil(len(list_of_columns)/2)) # since columns fixed at 2, no. of rows req = len(list)/2
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=[15,10]) # set subplots grid and figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns):
        ax[i].hist(dataframe[column], bins=20) # plot histogram of target values, bins default value 20
    for i, title in enumerate(list_of_titles):
        ax[i].set_title(title, size=15) # set hist title
    for i, label in enumerate(list_of_xlabels):
        ax[i].set_xlabel(label) # set xlabels
````

#### Plot and interpret boxplots

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:
- Only include variables of a similar scale
- Have clear labels for each variable
- Have appropriate titles and labels

````python
def subplot_box(dataframe, list_of_columns): # to plot at least TWO charts side by side for comparisons
    nrows = int(np.ceil(len(list_of_columns)/2)) # since columns fixed at 2, no. of rows req = len(list)/2
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=[15,(6*nrows)])
    titles = [] # create empty list for titles
    for col in list_of_columns: # iterate to append formatted titles
        titles.append("Boxplot of {} scores for {} in {}."\
                      .format(col.split("_")[-1].upper(), col.split("_")[0].upper(), col.split("_")[1])) # format titles
    y = 0
    for _, column in enumerate(list_of_columns):
        i, j = divmod(y, 2) # creates tuple
        ax[i,j].boxplot(dataframe[column]) # plot boxplot of target values
        ax[i,j].set_ylabel("Test Scores") # set ylabels
        ax[i,j].set_xlabel("{}".format(column.split("_")[-1].upper())) # set xlabels
        y += 1 # set up y to +1 with every loop
    y = 0
    for _, title in enumerate(titles):
        i, j = divmod(y, 2)
        ax[i,j].set_title(title, size=15) # set scatterplot title
        y += 1 # set up y to +1 with every loop
````

````python
dataframe = final
list_of_columns = ["sat_2017_participation", "act_2017_participation",
                   "sat_2018_participation", "act_2018_participation"] # boxplot for all SAT and ACT participation rates
subplot_box(dataframe=dataframe, list_of_columns=list_of_columns);
````

##### Does it make sense to conduct statistical inference given these data specifically? 

Why or why not?

*(think about granularity, aggregation, the relationships between populations size & rates...consider the actually populations these data describe in answering this question)*

Answer:

No, it does not make sense to conduct statistical inference. We have lost most, if not all, granularity in our data when we are given the average participation rates of each of the 51 States. If we were to consider these average rates as sample statistics, evaluating the mean of all 51 average rates gives us already the population mean (parameter) without needing to infer.

A possible scenario where it makes sense to conduct statistical inference is if we were to use this data as samples to infer parameters about the world at large as the population. For example, to infer the global participation rate of SAT and ACT. But even then, because these samples came wholly from USA, these samples are not random and are therefore terrible samples to be using for statistical inference. So even in this scenario, it does not make much sense to use these specific data.
